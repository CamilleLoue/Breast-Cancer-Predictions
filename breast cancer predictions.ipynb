## Model Training

```python
X = df_cancer.drop(['target'], axis=1)
y = df_cancer['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

#### Support Vector Machine

```python
from sklearn.svm import SVC

svc_model = SVC()
svc_model.fit(X_train, y_train)
```

## Model Evaluation

```python
y_predict = svc_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_predict))
```

```python
print(confusion_matrix(y_test,y_predict))
```

```python
sns.heatmap(confusion_matrix(y_test,y_predict), annot=True)
```

## Model Improvement (with Data Normalization)

#### Data normalization of the train dataset

```python
# We can use StandardScaler from sklearn but here we will do by-hand step-by-step
min_train = X_train.min()
min_train
```

```python
range_train = (X_train - min_train).max()
range_train
```

```python
X_train_scaled = (X_train - min_train)/range_train
X_train_scaled
```

```python
# Visualization without normalization
sns.scatterplot(x='mean area', y='mean smoothness', data=X_train, hue=y_train)
```

```python
# Visualization with normalization
sns.scatterplot(x='mean area', y='mean smoothness', data=X_train_scaled, hue=y_train)
```

#### Data normalisation of the test dataset

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test)/range_test
```

### Training, predicting & evaluating the model

```python
from sklearn.svm import SVC

svc_model = SVC()
svc_model.fit(X_train_scaled,y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_predict))
```

```python
sns.heatmap(confusion_matrix(y_test,y_predict), annot=True)
```

## Model Improvement - Part 2 (with Parameters Optimization : C & gama)

```python
 param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,verbose=4,refit=True)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_estimator_
```

```python
grid_predict = grid.predict(X_test_scaled)
```

```python
print(classification_report(y_test,grid_predict))
```

```python
sns.heatmap(confusion_matrix(y_test,grid_predict), annot=True)
```
